## 四、Plotting Tools

### 1 矩阵绘图 Scatter Matrix Plot

```python
from pandas.tools.plotting import scatter_matrix
df = pd.DataFrame(np.random.randn(1000, 4), columns=['a', 'b', 'c', 'd'])
df[:5]
```

```python
scatter_matrix(df, alpha=0.2, figsize=(6, 6), diagonal='kde')
```

### 2 Density Plot

```python
ser = pd.Series(np.random.randn(1000))
ser[:5]
```

```python
ser.plot.kde()
```

### 3 Andrews Curves

```python
from pandas.tools.plotting import andrews_curves
from sklearn.datasets import load_iris

d = load_iris()
data = pd.DataFrame(d.data,columns = d.feature_names)
data.columns.name = 'feature_names'
data[:5]
```

### 4 Parallel Coordinates

```python
from pandas.tools.plotting import parallel_coordinates

plt.figure()
#parallel_coordinates(data, 'sepal length')
```

```python
from pandas.tools.plotting import lag_plot

plt.figure()
data = pd.Series(0.1 * np.random.rand(1000) +
     0.9 * np.sin(np.linspace(-99 * np.pi, 99 * np.pi, num=1000)))
lag_plot(data)
```

### 5 Autocorrelation Plot

```python
from pandas.tools.plotting import autocorrelation_plot

plt.figure()
data = pd.Series(0.7 * np.random.rand(1000) +
       0.3 * np.sin(np.linspace(-9 * np.pi, 9 * np.pi, num=1000)))
autocorrelation_plot(data)
```

### 6 Bootstrap Plot

```python
from pandas.tools.plotting import bootstrap_plot

data = pd.Series(np.random.rand(1000))
bootstrap_plot(data, size=50, samples=500, color='grey')
```
