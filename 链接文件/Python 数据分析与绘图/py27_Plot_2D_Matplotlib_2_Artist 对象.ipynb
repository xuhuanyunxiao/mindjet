```python
plt.plot([1,2,3],[4,5,6])
plt.show()
axis = plt.gca().xaxis
```

```python
print axis.get_ticklocs() # 获得axis对象的刻度位置列表
print axis.get_ticklabels() # 获得刻度标签列表
print axis.get_ticklines() # 获得X轴上表示主刻度线的列表
print axis.get_ticklines(minor=True) # 副刻度线列表
[x.get_text() for x in axis.get_ticklabels()] # 获得刻度的文本字符串
```

```python
# 设置各种属性
for label in axis.get_ticklabels():
    label.set_color('red')
    label.set_rotation(45)
    label.set_fontsize(16)
    
for line in axis.get_ticklines():
    line.set_color('green')
    line.set_markersize(25)
    line.set_markeredgewidth(3)
```

```python
# 自定义坐标轴的刻度和文字

import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator,FuncFormatter
import numpy as np

x = np.arange(0,4*np.pi,0.01)
y = np.sin(x)
plt.figure(figsize=(8,4))
plt.plot(x,y)
ax = plt.gca()

def pi_formatter(x,pos): # 比较啰嗦地将数值转换为以 pi/4 为单位的刻度文本
    m = np.round(x/(np.pi/4))
    n = 4
    while m!=0 and m%2==0:m,n=m//2,n//2
    if m == 0:
        return '0'
    elif m == 1 and n ==1:
        return '$\pi$'
    elif n == 1:
        return r'$%d \pi$' % m
    elif m == 1:
        return r'$\frac{\pi}{%d}$' % n
    return r'$\frac{%d \pi}{%d}$' %(m,n)

plt.ylim(-1.5,1.5) # 设置坐标轴范围
plt.xlim(0,np.max(x))

plt.subplots_adjust(bottom = 0.15) # 设置图的底边距
plt.grid() # 开启网格

ax.xaxis.set_major_locator(MultipleLocator(np.pi/4)) # 主刻度为 pi/4
ax.xaxis.set_major_formatter(FuncFormatter(pi_formatter)) # 主刻度文本用 pi_formatter 函数计算
ax.xaxis.set_minor_locator(MultipleLocator(np.pi/20)) # 副刻度为 pi/20

for tick in ax.xaxis.get_major_ticks(): # 设置刻度文本的大小
    tick.label1.set_fontsize(16)
    
plt.show()
```
