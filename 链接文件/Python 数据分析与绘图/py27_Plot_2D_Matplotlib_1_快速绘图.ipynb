### 1 pyplot 

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,1000)
y = np.sin(x)
z = np.cos(x**2)

plt.figure(figsize = (8,4)) # 创建Figure对象
# figsize指定宽度和高度，单位英寸；dpi指定分辨率，即没英寸所表示的像素数，默认为80（显示）和100（保存） 

plt.plot(x,y,label='$sin(x)$',color='red',linewidth=2)
plt.plot(x,z,'b--',label='$cos(x^2)$')

plt.xlabel('Time(s)')
plt.ylabel('Volt')
plt.title('PyPlot Fist Example')
plt.ylim(-1.2,1.2)
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(1) # 图表1
plt.figure(2) # 图表2
ax1 = plt.subplot(211) # 图表2中的子图1
ax2 = plt.subplot(212) # 图表2中的子图2

x = np.linspace(0,3,100)
for i in xrange(5):
    plt.figure(1) # 图表1
    plt.plot(x,np.exp(i*x/3))
    plt.sca(ax1) # 图表2中的子图1
    plt.plot(x,np.sin(i*x))
    plt.sca(ax2) # 图表2中的子图2
    plt.plot(x,np.cos(i*x))
    
plt.show()
```

```python
# 显示所有的中文字体
from matplotlib.font_manager import fontManager
import matplotlib.pyplot as plt
import os

fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(111)
plt.subplots_adjust(0,0,1,1,0,0)
plt.xticks([])
plt.xticks([])
x,y = 0.05,0.08
fonts = [font.name for font in fontManager.ttflist if
        os.path.exists(font.fname) and os.stat(font.fname).st_size>1e6] # os.stat() 获取文件大小
font = set(fonts)
dy = (1.0-y)/(len(fonts)/4 + (len(fonts)%4!=0))

for font in fonts:
    t = ax.text(x,y,u'中文字体',{'fontname':font,'fontsize':14},transform=ax.transAxes)
    ax.text(x,y-dy/2,font,transform=ax.transAxes)
    x += 0.25
    if x >= 1.0:
        y += dy
        x = 0.05

plt.show() 
```
